midiファイルから各音符に対応するIDを出力するプログラム。とりあえずdurationは無視する。

```python
from music21 import converter
```

```python
filepath = '/path/to/midifiles'
```

```python
UNK = '<UNK>'
PAD = '<PAD>'
vocab = {PAD:0, UNK:1}
```

```python
# 出現する全ての音符のリスト
notes = list(set(n.nameWithOctave for n in converter.parse(filepath).flat.pitches))
```

```python
for note in notes:
  vocab[note] = len(vocab)
```

```python
# 確認
vocab
```

```python
# 一連の操作を定義
def build_vocabulary(filepath):
  """
  file: str, 入力midiファイルのパス
  """
  UNK = '<UNK>'
  PAD = '<PAD>'
  vocab = {PAD:0, UNK:1}
  notes = list(set(n.nameWithOctave for n in converter.parse(filepath).flat.pitches))
  for note in notes:
    vocab[note] = len(vocab)
  return vocab
```
